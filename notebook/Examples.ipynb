```python
import pyhml
hmlparser = pyhml.HmlParser()
hml_file = "hml_1_0_1-example7-ngsFull.xml"
```

```python
# Parse HML file into HML object
hml = hmlparser.parse(hml_file)
```

```python
# Convert HML object to pandas dataframe
hml_df = hml.toPandas()
hml_df
```

```python
for subject in hml.sample:
    print(subject)
```

```python
for subject in hml.sample:
    for typing in subject.typing:
        for allele_assignment in typing.allele_assignment:
            print(subject.id, allele_assignment.glstring)
```
